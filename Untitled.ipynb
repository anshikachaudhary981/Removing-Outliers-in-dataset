```python
#removing outlier from z score formula
outlier=[]
def remove(data):
    threshold=3
    mean=np.mean(data)
    std=np.std(data)
    
    for i in data:
        z_score=(i - mean)/std
        if np.abs(z_score)>threshold:
            outlier.append(i)
    return outlier
```

```python
out=remove(cars['price'])
out
```

```python
cars1=np.min(out)
cars1
```

```python
values=[]
def value(pp):
    for y in pp:
        if y <55600:
            values.append(y)
    return values
```

```python
a=value(cars['price'])
np.max(a)
```

```python
#After removing outliers 
# we are using our function because our data has remaining outliers
out2=remove(a)
value2=value(out2)
value2
plt.boxplot(value2)
```

```python
# we can see that we have some outliers also after removing 2 times
out3=remove(value2)
out3
```

```python
value3=value(out3)
value3
```

```python
plt.boxplot(value3)
```
